```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

pd.set_option('mode.chained_assignment', None)
```

```python
cameras_and_crashes= "Output Data/cameras_and_crashes.csv"
cameras_and_crashes_clean= "Output Data/cameras_and_crashes_clean.csv"
violations_cameras_clean= "Output Data/violations_cameras_clean.csv"
```

```python
cameras_and_crashes_df= pd.read_csv(cameras_and_crashes)
cameras_and_crashes_clean_df= pd.read_csv(cameras_and_crashes_clean)
violations_cameras_clean_df= pd.read_csv(violations_cameras_clean)
cameras_and_crashes_clean_df.head()
```

```python
x_values= cameras_and_crashes_clean_df['TOTAL CAMERAS LIVE']
y_values= cameras_and_crashes_clean_df['TOTAL CRASHES']

plt.title('TOTAL CAMERAS VS CRASHES')

plt.xlabel('TOTAL CAMERAS LIVE')
plt.ylabel('TOTAL CRASHES')
plt.scatter(x_values, y_values) 

plt.show()
```

```python
#How does the number of speed cameras correlate with the amount of car crashes?

#linear

x_values= cameras_and_crashes_clean_df["TOTAL CAMERAS LIVE"]
y_values= cameras_and_crashes_clean_df["TOTAL CRASHES"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.scatter(x_values, y_values) 
plt.title("TOTAL CAMERAS VS CRASHES")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(125,20000),fontsize=13,color="red")
plt.xlabel('TOTAL CAMERAS LIVE')
plt.ylabel('TOTAL CRASHES')

print(f"The r-squared is: {rvalue**2}")
print("How does the number of speed cameras correlate with the amount of car crashes?")
print("The regression shows a medium correlation.")
predicted_crashes= slope * 200 + intercept
print(f"For 2022, if the City of Chicago added enough cameras to get to 200 in total, we are predicting that the number of crashes would be {predicted_crashes}")
print("As the total number of cameras increases, the number of crashes increases.")

plt.show()
```

```python
#How does the number of speed cameras correlate with the amount of car violations?

#2022 Prediction of violations (using linear regression)?

x_values= violations_cameras_clean_df["TOTAL CAMERAS LIVE"]
y_values= violations_cameras_clean_df["TOTAL VIOLATIONS"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.title('TOTAL CAMERAS VS VIOLATIONS')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(130,2000),fontsize=12,color="red")
plt.xlabel('TOTAL CAMERAS LIVE')
plt.ylabel('TOTAL VIOLATIONS')
plt.scatter(x_values, y_values) 

print(f"The r-squared is: {rvalue**2}")
print("How does the number of speed cameras correlate with the amount of car violations?")

print("The regression shows a weak correlation.")
print("As the total number of cameras increases, the number of violations increases.")
predicted_violations= slope * 200 + intercept
print(f"For 2022, if the City of Chicago added enough cameras to get to 200 in total, we are predicting that the number of violations would be {predicted_violations}")

plt.show()
```
